### 程式實力ch3_24.py:從地球到月球約是384440公里，假設火箭的速度是一馬赫，設計一個程式計算需要多少天、多少小時才可以抵達月球。這個程式省略分鐘數。

```python
#CH3_24.py
dist= 384440                    #地球到月球的距離
speed= 1225                     #馬赫速度每小時1225公里
total_hours = dist // speed     #計算小時數
days = total_hours // 24        #商 = 計算天數
hours = total_hours % 24        #餘 =計算小時數
print("總共需要多少天")
print(days)
print("小時數")
print(hours)
```

### 從地球到上海約是10000公里，假設火箭的速度是一馬赫，設計一個程式計算需要多少天、多少小時才可以抵達上海。

```python
#CH3_24.py
dist= 10000
speed= 1225
total_hours = dist // speed
days = total_hours // 24
hours = total_hours % 24
print("總共需要多少天")
print(days)
print("小時數")
print(hours)
```

### 程式實力ch4_20.py:請輸入華氏溫度，這個程式會輸出攝氏溫度

```python
#ch5_4.py
print("奇數偶數判斷")
num = input("請輸入任意整數值:")
rem = int(num) % 2
if (rem ==0):
    print("%d 是偶數" % int(num))
else:
    print("%d 是奇數" % int(num))
```
