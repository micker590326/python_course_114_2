```python
# ch5_5. py
print("計算最終成績")
score = input("請輸入分數")
sc = int(score)
if (sc >= 90):
    print(" A")
elif (sc >= 80):
    print(" B")
elif (sc >= 70):
    print(" C")
elif (sc >= 60):
    print(" D")
else:
    print(" F")
```

練習 5-1：出生年 year 先減去 1900・再計算 rem=year % 12。
rem==0是鼠・rem=1是牛・...rem==11是屬豬
請寫一程式，可輸入西元年，輸出生肖

```python
print("輸出生肖")
year = input("age")
age = int(year)
rem=(age-1900)%12
if (rem== 0):
    print(" 鼠")
elif (rem== 1):
    print(" 牛")
elif (rem== 2):
    print(" 虎")
elif (rem== 3):
    print(" 吐")
elif (rem== 4):
    print(" 龍")
elif (rem== 5):
    print(" 蛇")
elif (rem== 6):
    print(" 馬")
elif (rem== 7):
    print(" 楊")
elif (rem== 8):
    print(" 侯")
elif (rem== 9):
    print(" 機")
elif (rem== 10):
    print(" 狗")
else:
    print(" 朱")
```
